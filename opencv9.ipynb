Histogram Analysis & Comparison

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Step 1: Read image
img = cv2.imread('/content/bougainvillea_00002.jpg')

if img is None:
    print("Error: Image not loaded")
else:
    # Step 2: Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Step 3: Histogram Equalization (grayscale)
    equalized = cv2.equalizeHist(gray)

    # Step 4: Display images
    print("Original Image")
    cv2_imshow(img)

    print("Grayscale Image")
    cv2_imshow(gray)

    print("Equalized Image")
    cv2_imshow(equalized)

    # Step 5: Plot histograms

    # 🔹 Grayscale histogram (single intensity)
    plt.figure()
    plt.title("Grayscale Histogram")
    plt.hist(gray.ravel(), bins=256, range=[0,256])
    plt.show()

    # 🔹 Equalized histogram
    plt.figure()
    plt.title("Equalized Histogram")
    plt.hist(equalized.ravel(), bins=256, range=[0,256])
    plt.show()

    # 🔹 Color histograms (B, G, R)
    colors = ('b', 'g', 'r')
    plt.figure()
    plt.title("Color Histogram")

    for i, col in enumerate(colors):
        hist = cv2.calcHist([img], [i], None, [256], [0,256])
        plt.plot(hist, color=col)

    plt.xlim([0,256])
    plt.show()
```
