5. Save an Image from Video

```python
import cv2
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture('/content/1.mp4')

frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    # Show only first 3 frames
    if frame_count < 1:
        print(f"Frame {frame_count}")
        cv2_imshow(frame)

    frame_count += 1

    if frame_count >= 1:
        break

cap.release()
```
